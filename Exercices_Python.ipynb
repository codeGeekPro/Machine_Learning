```python
hauteur = int(input("Saisir le nombre des lignes " ) )
nombre_espace = hauteur - 1
nombre_etoile = 1
for i in range (0,hauteur):
    print(nombre_espace* " " + nombre_etoile* "∗" )
    nombre_espace -= 1
    nombre_etoile += 2
```
