Pour **renverser une image** reprÃ©sentÃ©e sous forme de tableau NumPy en Python, tu peux utiliser les **opÃ©rations sur les indices NumPy**. Voici plusieurs mÃ©thodes selon le type de renversement souhaitÃ© :

### **1. Renverser l'image verticalement (haut â†” bas)**
Utilise `numpy.flipud()` ou `[::-1, :]` :

```python
import numpy as np
import matplotlib.pyplot as plt

# Charger une image en tant que tableau numpy
image = np.random.rand(100, 100, 3)  # Image RGB alÃ©atoire (100x100)

# Renverser verticalement
image_flip_v = np.flipud(image)  # ou image[::-1, :]

# Afficher l'image originale et renversÃ©e
fig, ax = plt.subplots(1, 2)
ax[0].imshow(image)
ax[0].set_title("Originale")
ax[1].imshow(image_flip_v)
ax[1].set_title("RenversÃ©e Verticalement")
plt.show()
```

---

### **2. Renverser l'image horizontalement (gauche â†” droite)**
Utilise `numpy.fliplr()` ou `[:, ::-1]` :

```python
# Renverser horizontalement
image_flip_h = np.fliplr(image)  # ou image[:, ::-1]

# Afficher
fig, ax = plt.subplots(1, 2)
ax[0].imshow(image)
ax[0].set_title("Originale")
ax[1].imshow(image_flip_h)
ax[1].set_title("RenversÃ©e Horizontalement")
plt.show()
```

---

### **3. Renverser l'image en diagonale (Transpose + Flip)**
Utilise `numpy.transpose()` + `numpy.flip()` :

```python
# Renverser en diagonale (Transpose puis flip)
image_flip_diag = np.flip(image.T, axis=1)  # ou image.transpose(1, 0, 2)[::-1, :]

# Afficher
fig, ax = plt.subplots(1, 2)
ax[0].imshow(image)
ax[0].set_title("Originale")
ax[1].imshow(image_flip_diag)
ax[1].set_title("RenversÃ©e en Diagonale")
plt.show()
```

---

### **4. Rotation de 180Â° (Inversement total)**
Utilise `numpy.flip()` deux fois ou `numpy.rot90(image, 2)` :

```python
# Rotation de 180Â° (Ã©quivalent Ã  flip horizontal + vertical)
image_rot_180 = np.flip(image)  # ou np.rot90(image, 2)

# Afficher
fig, ax = plt.subplots(1, 2)
ax[0].imshow(image)
ax[0].set_title("Originale")
ax[1].imshow(image_rot_180)
ax[1].set_title("RotÃ©e 180Â°")
plt.show()
```

---

### **RÃ©sumÃ© :**
| Transformation | MÃ©thode |
|--------------|---------|
| Renversement vertical | `np.flipud(image)` ou `image[::-1, :]` |
| Renversement horizontal | `np.fliplr(image)` ou `image[:, ::-1]` |
| Renversement diagonal | `image.T[::-1, :]` ou `np.flip(image.T, axis=1)` |
| Rotation 180Â° | `np.flip(image)` ou `np.rot90(image, 2)` |
