on='key' : Cet argument spécifie la colonne (ou les colonnes) sur laquelle vous souhaitez fusionner les deux DataFrames. Dans ce cas, la colonne est 
appelée key. Les deux DataFrames doivent avoir une colonne avec ce nom pour que la fusion soit possible.

how='outer' : Cet argument détermine le type de fusion que vous voulez effectuer. Il y a plusieurs options pour cet argument :

inner : Effectue une fusion interne. Seules les lignes avec des valeurs correspondantes dans les deux DataFrames seront conservées.

outer : Effectue une fusion externe. Toutes les lignes des deux DataFrames seront conservées, même s'il n'y a pas de correspondance dans l'autre 
DataFrame. Les valeurs manquantes seront remplacées par NaN.

left : Effectue une fusion gauche. Toutes les lignes du DataFrame de gauche (df1) seront conservées, et seulement les lignes correspondantes du
DataFrame de droite (df2) seront ajoutées.

right : Effectue une fusion droite. Toutes les lignes du DataFrame de droite (df2) seront conservées, et seulement les lignes correspondantes du 
DataFrame de gauche (df1) seront ajoutées.

Dans cet exemple, la fonction merge combine df1 et df2 en utilisant la colonne key comme clé de fusion, et inclut toutes les lignes des deux 
DataFrames, même si elles ne correspondent pas dans l'autre DataFrame (fusion externe).

# EXERCICE Trading Strategy , Turtle Strategy 
"""
Comment procéder :
1-) uTILISER LA FONCTION ROLLING POUR CALCULER :
 -MAX SUR LES 28 DERNIERS JOURS
 -mIN SUR LES 18 DERNIERS JOURS 

2-) le boolean indexing 
  - Si 'close'> max28 alors buy =1
  -si 'Close'<min28 alors sell =-1

  indice initialiser les colonnes 'buy ' et 'sell' avec des 0

bitcoin['Buy']=np.zeros(len(bitcoin))
bitcoin['Sell']=np.zeros(len(bitcoin))

"""

```python
data = pd.read_csv('historical_stock_data.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

data['SMA_50'] = data['Close'].rolling(window=50).mean()
data['SMA_200'] = data['Close'].rolling(window=200).mean()

data['Signal'] = 0
data['Signal'][50:] = np.where(data['SMA_50'][50:] > data['SMA_200'][50:], 1, 0)
data['Position'] = data['Signal'].diff()

plt.figure(figsize=(12,8))
plt.plot(data['Close'], label='Close Price')
plt.plot(data['SMA_50'], label='50-Day SMA')
plt.plot(data['SMA_200'], label='200-Day SMA')
plt.plot(data[data['Position'] == 1].index, data['SMA_50'][data['Position'] == 1], '^', markersize=10, color='g', label='Buy Signal')
plt.plot(data[data['Position'] == -1].index, data['SMA_50'][data['Position'] == -1], 'v', markersize=10, color='r', label='Sell Signal')
plt.title('Trading Strategy')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Lire les données historiques des actions à partir d'un fichier CSV
data = pd.read_csv('BTC-EUR.csv')

# Convertir la colonne 'Date' en type datetime
data['Date'] = pd.to_datetime(data['Date'])

# Définir la colonne 'Date' comme index
data.set_index('Date', inplace=True)

# Calculer la moyenne mobile sur 50 jours de la colonne 'Close'
data['SMA_50'] = data['Close'].rolling(window=50).mean()

# Calculer la moyenne mobile sur 200 jours de la colonne 'Close'
data['SMA_200'] = data['Close'].rolling(window=200).mean()

# Initialiser la colonne 'Signal' à 0
data['Signal'] = 0

# Générer des signaux d'achat (1) et de vente (0) à partir des moyennes mobiles
data['Signal'][50:] = np.where(data['SMA_50'][50:] > data['SMA_200'][50:], 1, 0)

# Calculer la différence pour obtenir les positions (1 pour achat, -1 pour vente)
data['Position'] = data['Signal'].diff()

# Créer une figure de la taille spécifiée
plt.figure(figsize=(12,8))

# Tracer les prix de clôture
plt.plot(data['Close'], label='Prix de clôture')

# Tracer la moyenne mobile sur 50 jours
plt.plot(data['SMA_50'], label='MMA sur 50 jours')

# Tracer la moyenne mobile sur 200 jours
plt.plot(data['SMA_200'], label='MMA sur 200 jours')

# Ajouter des marqueurs pour les signaux d'achat
plt.plot(data[data['Position'] == 1].index, data['SMA_50'][data['Position'] == 1], '^', markersize=10, color='g', label='Signal d\'achat')

# Ajouter des marqueurs pour les signaux de vente
plt.plot(data[data['Position'] == -1].index, data['SMA_50'][data['Position'] == -1], 'v', markersize=10, color='r', label='Signal de vente')

# Ajouter un titre au graphique
plt.title('Stratégie de trading')

# Ajouter un label à l'axe des x
plt.xlabel('Date')

# Ajouter un label à l'axe des y
plt.ylabel('Prix')

# Afficher la légende
plt.legend()

# Afficher le graphique
plt.show()
```

```python
# reprenons l'exercice précédent en utilisant maintenant une fonction
def category_ages(age):
    if age<=20 :
        return '<20 ans'
    elif (age>20) & (age<=30):
        return '20-30 ans'
    elif (age>30) & (age<=40):
        return '30-40 ans'
    else:
        return '+40 ans'
```
