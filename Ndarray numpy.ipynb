```python
import numpy as np
import matplotlib.pyplot as plt


# Charger l'image
img = np.asarray(face)

# Obtenir les dimensions de l'image
height, width, _ = img.shape

# Définir les limites pour le zoom central
start_row, end_row = height // 4, 3 * height // 4
start_col, end_col = width // 4, 3 * width // 4

# Extraire la sous-région
zoomed_img = img[start_row:end_row, start_col:end_col]

# Afficher l'image zoomée
plt.imshow(zoomed_img)
plt.show()
```

# 1-) Le module pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,20,100)
y=x**2
plt.plot(x,y, c='red', lw=1,ls='--')# amusons nous avec des paramètres de personnalisation
plt.show()
```

```python
# pour tracer un nuage de point , on utilise la fonction scatter
plt.scatter(x,y)
```

```python
# Pour bien faire on commence par 
plt.figure(figsize=(10,6))  # cette ligne nous crée une nouvelle fenêtre de travail
plt.title('figure 1')  # exemple de code pour afficher le titre
plt.xlabel('Axe des abscisses')  # on nomme les axes
plt.ylabel('axe des ordonnées')
plt.plot(x,y,label='quadratique')
plt.plot(x,x**3, label='cubique')
plt.legend()
plt.savefig('figuer_sauvegardée.png') # pour sauvegarder la figure sous forme de fichier png
```

```python
# on va apprendre à afficher plusieurs graphique sur la même figure
# avec la fonction plt.subplot(lignes,colonnes,position)
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.plot(x,y,c='green')
plt.subplot(2,2,2)
plt.plot(x,x**3,ls='--',label='cubique')
plt.legend()
plt.subplot(2,2,3)
plt.plot(x,np.log(x),label='Fonction logarithmique', c='red')
plt.legend()
plt.subplot(2,2,4)
plt.plot(x,np.exp(x),label='Fonction exponentielle',c='yellow')
plt.legend()

plt.savefig('Graphique.png')
```

```python
# Utilisons un peu la méthode orintée objet pour tracer les courbes
fig, ax= plt.subplots()  # Faire très attention quant à la syntaxe ici
ax.plot(x,y)
```

```python
fig, ax= plt.subplots(2,1,sharex=True)  # une petite particularité est d'avoir deux graphique le m*eme axe des abscisses
ax[0].plot(x,y)
ax[1].plot(x,np.cos(x))
plt.show()
```

```python
# correction de l'exercice
dataset={f"expérience{i}": np.random.randn(100) for i in range (4)}
```

```python
def graphique(dataset):
    n=len(dataset)
    plt.figure(figsize=(12,10))
    for k,i in zip(dataset.keys(),range(1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(dataset[k])
        plt.title(k)
    plt.show()
```

```python
# Application de la fonction
graphique (dataset)
```
